
## counts groupby publisher 


```python
uglydata= pd.read_csv('article_data_and_price_labeled_publisher.csv',index_col=False).drop('Unnamed: 0', axis = 1)
uglydata=pd.DataFrame(uglydata)
uglydata.head(40) 
```

```python
uglydata_author=pd.Series(uglydata.author)
uglydata_author.to_csv('uglydata_author.csv')

```

```python

ugly_read=pd.read_csv('uglydata_author.csv')


data = {'author': ugly_read.iloc[:,1]}

df = pd.DataFrame(data)
# df['author'] = df['author'].replace(r'http\S+', '', regex=True).replace(r'www\S+', '', regex=True)
mask = df['author'].str.contains('http|.com|.org|https|www.|@|.co|.net')
# df = df.drop(mask)
# df.dropna()
# df.head(60)
df = df[mask==False]
df.head(40)

```

```python
uglydata_publisher=pd.Series(uglydata.publisher)
uglydata_publisher.to_csv('uglydata_publisher.csv')
ugly_read_2=pd.read_csv('uglydata_publisher.csv')
data_2 = {'publisher': ugly_read_2.iloc[:,1]}

df_2 = pd.DataFrame(data_2)
author_and_publisher=pd.merge(df, df_2, right_index=True, left_index=True) 
author_and_publisher.head(50).reset_index()
```

```python
author_and_publisher['counts']=np.ones(len(author_and_publisher),np.int)
author_and_publisher=author_and_publisher.groupby(['author']).sum().sort_values('counts',ascending=True)
author_and_publisher
```
