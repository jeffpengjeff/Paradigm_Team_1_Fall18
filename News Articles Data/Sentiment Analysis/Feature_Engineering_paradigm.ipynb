### --cleaing the description part

```python
data_labeled_publisher_author = pd.read_csv('dataset_labeled_publisher_author.csv')
data_labeled_publisher_author.head(5) 
```

```python
#Clean the articles - Remove stopwords, remove punctuation, all lowercase
import re
for i in data_labeled_publisher_author .index:
    text = data_labeled_publisher_author .loc[i, 'description']
    if pd.isnull(text):
        pass
    else:
        text = re.sub(r"[^a-zA-Z]", " ", text)
        text = [word for word in text.split() if not word in eng_stopwords]
        text = (' '.join(text))
        text = text.lower()
        data_labeled_publisher_author.loc[i, 'description'] = text
```

```python
def gc_sentiment(text):  
    from google.cloud import language
    
    client = language.LanguageServiceClient()
    document = language.types.Document(
            content=text,
            type=language.enums.Document.Type.PLAIN_TEXT)
    annotations = client.analyze_sentiment(document=document)
    score = annotations.document_sentiment.score
    magnitude = annotations.document_sentiment.magnitude
    return score, magnitude
```

```python
from tqdm import tqdm # This is an awesome package for tracking for loops
gc_results = [gc_sentiment(row) for row in tqdm(contents_p1 , ncols = 100)]
gc_score, gc_magnitude = zip(*gc_results) # Unpacking the result into 2 lists
gc = list(zip(contents_p1 , gc_score, gc_magnitude))
columns = ['text', 'score', 'magnitude']
result = pd.DataFrame(gc, columns = columns)
```
